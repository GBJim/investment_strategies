```python
from investment_strategies import Strategy
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def render_histogram(k_frequency):
    X = np.arange(len(k_frequency)) + 1
    Y = k_frequency
    f, ax = plt.subplots(1)

    ax.plot(X, Y)
    ax.set_ylim(ymin=0)
    plt.show(f)
```

```python
degree = 1
goal = 1000
repetitions=10000
```

```python
#Risky strategy
a = 1
b = 1
parameters = [a,b]
s = Strategy(degree, parameters=[a,b])
k_frequency = s.get_k_frequency(goal, repetitions)
render_histogram(k_frequency)
```

```python
#Uniform distribution strategy

a = 1
b = 0
parameters = [a,b]
s = Strategy(degree, parameters=[a,b])
k_frequency = s.get_k_frequency(goal, repetitions)
render_histogram(k_frequency)



```

```python
#Safe strategy

a = 0
b = -1
parameters = [a,b]
s = Strategy(degree, parameters=[a,b])
k_frequency = s.get_k_frequency(goal, repetitions)
render_histogram(k_frequency)

```
