```python
def render_steps_histogram(steps):
    f, ax = plt.subplots(1)
    ax.hist(steps,bins=500, edgecolor='none')
    plt.xlabel("Steps")
    plt.ylabel("Frequency")
    plt.title("Distribution of Steps")
    plt.xlim([1, 5000])
    plt.ylim([0,2500])
    plt.show()
```
